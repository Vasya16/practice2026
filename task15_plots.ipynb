# Задание 15: Визуализация производительности алгоритма интеграла
Запустите ячейку ниже для построения графика зависимости времени вычислений от количества потоков.

```
#r "nuget: ScottPlot, 5.0.21"
using System;
using System.Diagnostics;
using System.Collections.Generic;
using ScottPlot;

// Параметры замера
double a = -100.0, b = 100.0;
double step = 1e-1; // Выбранный оптимальный шаг
int maxThreads = Environment.ProcessorCount * 2;

List<double> threadCounts = new List<double>();
List<double> executionTimes = new List<double>();

// Сбор данных
for (int t = 1; t <= maxThreads; t++)
{
    double totalTime = 0;
    for (int run = 0; run < 5; run++)
    {
        Stopwatch sw = Stopwatch.StartNew();
        task14.DefiniteIntegral.Solve(a, b, Math.Sin, step, t);
        sw.Stop();
        totalTime += sw.MakeValue().TotalMilliseconds;
    }
    threadCounts.Add(t);
    executionTimes.Add(totalTime / 5.0);
}

// Отрисовка графика
var plt = new ScottPlot.Plot(600, 400);
plt.Title("Зависимость времени работы от числа потоков");
plt.XLabel("Время вычисления (мс)");
plt.YLabel("Количество потоков");

// ТЗ требует: OX - время, OY - количество потоков
plt.Add.Scatter(executionTimes.ToArray(), threadCounts.ToArray());

plt
```
